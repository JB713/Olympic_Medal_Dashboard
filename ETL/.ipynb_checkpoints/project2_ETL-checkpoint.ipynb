```python
def country_fun(col):
    return str(col)[0:2] + '_c_' + str(col)[-1]
country = dic
country['country_id'] = dic.Code.apply(country_fun)
country
```

```python
country = country[['country_id','Country','Code','Population','GDP']]
```

```python
country
```

```python
athlete = game[['Athlete','Gender','Country']].drop_duplicates()
#athlete[athlete[]]
def athlete_fun(df):
    return str(df[0]).replace(", ","-") + '_c_' + str(df[1])[0:2] + '_d_' + str(df[2])[0:5]

athlete['Athlete_id'] = athlete.apply(athlete_fun, axis =1)
athlete = pd.merge(athlete, country, left_on='Country', right_on='Code',how = 'left')[['Athlete','Gender','Athlete_id','country_id']]
athlete = athlete[['country_id','Athlete','Athlete_id','Gender']]
```

```python
#aa = athlete['Athlete_id'=='CH_c_e_d_GRE']
#athlete[athlete['Athlete_id']=='PERSA_c_e_d_GRE']
#= athlete['Athlete_id'].drop.duplicates()
athlete
```

```python
game1=game[['Year','Sport','Discipline','Event','Season','City']].drop_duplicates()
def game_fun(df):
    return str(df[0]) + str(df[1]) + '_c_' + str(df[2]) + '_d_' + str(df[3]).replace(" ",".") + str(df[4])
game1['event_id'] = game1.apply(game_fun, axis =1)
game1
```

```python
game1 = game1[['event_id','Year','City','Sport','Discipline','Event','Season']]
```

```python
game1.info()
```

```python
medal = game['Medal'].drop_duplicates()
medal_df = pd.DataFrame(medal)
```

```python
medal_df
```

```python
def medal_fun(col):
    return str(col)[0:2] + '_c_' + str(col)[-1]
#medal = game['Medal'].drop_duplicates()
medal_df['medal_id'] = medal_df.Medal.apply(medal_fun)
medal_df
```

```python
medal_df = medal_df[['medal_id','Medal']]
```

```python
medal_df
```

```python
game['event_id'] = game[['Year','Sport','Discipline','Event','Season']].apply(game_fun, axis =1)
game['Athlete_id'] = game[['Athlete','Gender','Country']].apply(athlete_fun, axis =1)
game['medal_id'] = game['Medal'].apply(medal_fun)
game
```

```python
master = game[['event_id','Athlete_id','medal_id']]
master
```

```python
connection_string = f'{username}:{password}@localhost:5432/olympics2'
engine = create_engine(f'postgresql://{connection_string}')
```

```python
country.to_sql(name='country', con=engine, if_exists='append', index=False)
```

```python
medal_df.to_sql(name='medal', con=engine, if_exists='append', index=False)
```

```python
master.to_sql(name='master', con=engine, if_exists='append', index=False)
```
